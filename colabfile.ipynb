```python
## do not run it
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import pandas as pd
from keras.utils import image_dataset_from_directory
from keras.models import Sequential
from keras import layers
from keras.callbacks import EarlyStopping
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

import os
os.makedirs("figures", exist_ok=True)

# ### Loading Data
main_path = '/content/drive/MyDrive/MRL_Project/mrl_eye/data/'
img_size = (84, 84)
batch_size = 64

# Train dataset
train_data = image_dataset_from_directory(
    main_path + 'train',
    image_size=img_size,
    batch_size=batch_size,
    shuffle=True,
    seed=42
)

# Validation dataset
val_data = image_dataset_from_directory(
    main_path + 'val',
    image_size=img_size,
    batch_size=batch_size,
    shuffle=False
)

# Test dataset (Using 'val' since 'test' folder doesn't exist)
test_data = image_dataset_from_directory(
    main_path + 'val',
    image_size=img_size,
    batch_size=batch_size,
    shuffle=True
)

# 1. Grab Class info FIRST before shrinking
classes = train_data.class_names
num_classes = len(classes)
print(f'Dataset has the following classes: {classes}')

# 2. QUICK TRAINING MODE: Shrink the dataset so it trains fast!
print("Applying Quick Training Mode...")
train_data = train_data.take(156) # ~10,000 images
val_data = val_data.take(31)      # ~2,000 images
test_data = test_data.take(31)    # ~2,000 images


# ### Data Visualization
# Merging data for visualization
all_data = train_data.concatenate(val_data).concatenate(test_data)

images_list = []
labels_list = []

for x_batch, y_batch in all_data:
    images_list.append(x_batch.numpy())
    labels_list.append(y_batch.numpy())

images = np.vstack(images_list)
labels = np.hstack(labels_list)

# Show one image from each class
plt.figure()
for i, class_name in enumerate(classes):
    class_indices = np.where(labels == i)[0]
    idx = np.random.choice(class_indices)
    plt.subplot(1, num_classes, i + 1)
    plt.imshow(images[idx].astype('uint8'))
    plt.title(class_name)
    plt.axis('off')

plt.savefig("figures/examples.pdf", bbox_inches="tight")
plt.show()

# Histogram
plt.figure()
plt.hist(labels, bins=[-0.25, 0.25, 0.75, 1.25])
plt.xticks([0, 1], ['Awake', 'Sleepy'])
plt.title('Data Distribution by Classes')
plt.xlabel('Classes')
plt.ylabel('Number of Samples')
plt.savefig("figures/histogram.pdf", bbox_inches="tight")
plt.show()


# ### Data Augmentation
data_augmentation = Sequential([
    layers.Input(shape=(img_size[0], img_size[1], 3)),
    layers.RandomRotation(0.25),
    layers.RandomFlip('horizontal'),
    layers.RandomZoom(0.1),
])


# ### Training the Convolutional Neural Network
def cnn_model(num_classes):
    model = Sequential([
        data_augmentation,
        layers.Conv2D(16, 3, activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(),

        layers.Conv2D(32, 3, activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(),

        layers.Conv2D(64, 3, activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(),

        layers.Dropout(0.5),
        layers.Flatten(),

        layers.Dense(64, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

model = cnn_model(num_classes)
model.summary()

# Early stopping to halt training when it stops improving
es = EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True, verbose=1)

history = model.fit(train_data,
                    epochs=50,
                    validation_data=val_data,
                    callbacks=[es])

# Training curves
plt.figure()
plt.subplot(121)
plt.plot(history.history['accuracy'], label='train')
plt.plot(history.history['val_accuracy'], label='val')
plt.xlabel('Epochs')
plt.title('Accuracy')
plt.legend()

plt.subplot(122)
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='val')
plt.title('Loss')
plt.xlabel('Epochs')
plt.legend()
plt.savefig("figures/training_curves.pdf", bbox_inches="tight")
plt.show()


# ### Confusion Matrix for Test Set
y_true_test = np.array([])
y_pred_test = np.array([])

for img, lab in test_data:
    y_true_test = np.concatenate((y_true_test, lab.numpy()))
    y_pred_prob_test = model.predict(img, verbose=0)
    y_pred_test = np.concatenate((y_pred_test, np.round(y_pred_prob_test).reshape(-1,)))

print(f"Test Model Accuracy is: {100 * accuracy_score(y_true_test, y_pred_test):.2f}%")

cm_test = confusion_matrix(y_true_test, y_pred_test)
ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=classes).plot()
plt.title("Confusion Matrix - Test Data")
plt.savefig("figures/confusion_matrix_test.pdf", bbox_inches="tight")
plt.show()


# ### Confusion Matrix for Train Set
y_true_train = np.array([])
y_pred_train = np.array([])

for img, lab in train_data:
    y_true_train = np.concatenate((y_true_train, lab.numpy()))
    y_pred_prob_train = model.predict(img, verbose=0)
    y_pred_train = np.concatenate((y_pred_train, np.round(y_pred_prob_train).reshape(-1,)))

print(f"Train Model Accuracy is: {100 * accuracy_score(y_true_train, y_pred_train):.2f}%")

cm_train = confusion_matrix(y_true_train, y_pred_train)
ConfusionMatrixDisplay(confusion_matrix=cm_train, display_labels=classes).plot()
plt.title("Confusion Matrix - Train Data")
plt.savefig("figures/confusion_matrix_train.pdf", bbox_inches="tight")
plt.show()


# ### Final Metrics
TN, FP, FN, TP = cm_test.ravel()

ACC = (TN + TP) / np.sum(cm_test)
print(f"Accuracy: {ACC:.4f}")

P = TP / (TP + FP) if (TP + FP) != 0 else 0
print(f"Precision: {P:.4f}")

R = TP / (TP + FN) if (TP + FN) != 0 else 0
print(f"Recall: {R:.4f}")

S = TN / (TN + FP) if (TN + FP) != 0 else 0
print(f"Specificity: {S:.4f}")

F1 = 2 * P * R / (P + R) if (P + R) != 0 else 0
print(f"F1-score: {F1:.4f}")

BA = (S + R) / 2
print(f"Balanced accuracy: {BA:.4f}")


# ### Visualization of Correct vs Incorrect Predictions
images_list_test = []
labels_list_test = []

for img_batch, lab_batch in test_data:
    images_list_test.append(img_batch.numpy())
    labels_list_test.append(lab_batch.numpy())

images_test = np.vstack(images_list_test)
labels_test = np.hstack(labels_list_test)

correct_prediction = (y_true_test == y_pred_test)

idx_correct = np.where(correct_prediction == 1)[0]
idx_incorrect = np.where(correct_prediction == 0)[0]

print(f"Total correct predictions: {len(idx_correct)}")
print(f"Total incorrect predictions: {len(idx_incorrect)}")

idx_good = np.random.choice(idx_correct, size=min(5, len(idx_correct)), replace=False)
idx_bad = np.random.choice(idx_incorrect, size=min(5, len(idx_incorrect)), replace=False)

plt.figure(figsize=(12, 6))

for i in range(len(idx_good)):
    plt.subplot(2, 5, i+1)
    plt.imshow(images_test[idx_good[i]].astype('uint8'))
    plt.title("Class: \n" + classes[labels_test[idx_good[i]]])
    plt.axis('off')

for i in range(len(idx_bad)):
    plt.subplot(2, 5, i + len(idx_good) + 1)
    plt.imshow(images_test[idx_bad[i]].astype('uint8'))
    plt.title("Class: \n" + classes[labels_test[idx_bad[i]]])
    plt.axis('off')

plt.savefig("figures/classification_examples.pdf", bbox_inches="tight")
plt.show()
```

```python
##do not run it
import tensorflow as tf
from tensorflow.keras import layers, Sequential
from tensorflow.keras.utils import image_dataset_from_directory

# 1. Setup paths and size (MobileNet loves 96x96)
local_path = '/content/local_data/'
img_size = (96, 96)
batch_size = 32

# 2. Load data from the FAST local drive
train_data_fast = image_dataset_from_directory(local_path + 'train', image_size=img_size, batch_size=batch_size, shuffle=True)
val_data_fast = image_dataset_from_directory(local_path + 'val', image_size=img_size, batch_size=batch_size, shuffle=True)

# 3. Load the pre-trained "Super Brain"
base_model = tf.keras.applications.MobileNetV2(input_shape=(96, 96, 3), include_top=False, weights='imagenet')
base_model.trainable = False # We keep the general knowledge frozen

# 4. Build the final model
model_mob = Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dropout(0.2),
    layers.Dense(1, activation='sigmoid')
])

model_mob.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 5. Fast Training (Take a slice of 10,000 images)
print("Starting the rocket ship training...")
history_mob = model_mob.fit(train_data_fast.take(312), epochs=5, validation_data=val_data_fast.take(62))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score

# 1. ACCURACY & LOSS GRAPHS
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history_mob.history['accuracy'], label='Train Accuracy')
plt.plot(history_mob.history['val_accuracy'], label='Val Accuracy')
plt.title('MobileNet Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history_mob.history['loss'], label='Train Loss')
plt.plot(history_mob.history['val_loss'], label='Val Loss')
plt.title('MobileNet Loss')
plt.legend()
plt.show()

# 2. CALCULATE PREDICTIONS FOR BOTH SETS
def get_all_preds(dataset, steps):
    y_true, y_pred, images = [], [], []
    for img, lab in dataset.take(steps):
        images.extend(img.numpy())
        y_true.extend(lab.numpy())
        p = model_mob.predict(img, verbose=0)
        y_pred.extend(np.round(p).reshape(-1))
    return np.array(y_true), np.array(y_pred), np.array(images)

print("Gathering results (this takes a moment)...")
y_true_train, y_pred_train, _ = get_all_preds(train_data_fast, 312)
y_true_test, y_pred_test, test_imgs = get_all_preds(val_data_fast, 62)

# 3. PRINT ACCURACIES
train_acc = accuracy_score(y_true_train, y_pred_train)
test_acc = accuracy_score(y_true_test, y_pred_test)
print(f"\nTrain Model Accuracy is: {train_acc*100:.2f}%")
print(f"Test Model Accuracy is: {test_acc*100:.2f}%")

# 4. CONFUSION MATRICES
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
cm_train = confusion_matrix(y_true_train, y_pred_train)
ConfusionMatrixDisplay(cm_train, display_labels=classes).plot(ax=ax[0], cmap='Blues')
ax[0].set_title("Confusion Matrix - Train Data")

cm_test = confusion_matrix(y_true_test, y_pred_test)
ConfusionMatrixDisplay(cm_test, display_labels=classes).plot(ax=ax[1], cmap='Greens')
ax[1].set_title("Confusion Matrix - Test Data")
plt.show()

# 5. DETAILED METRICS BLOCK
TN, FP, FN, TP = cm_test.ravel()
prec = TP / (TP + FP) if (TP + FP) != 0 else 0
rec = TP / (TP + FN) if (TP + FN) != 0 else 0
spec = TN / (TN + FP) if (TN + FP) != 0 else 0
f1 = 2 * (prec * rec) / (prec + rec) if (prec + rec) != 0 else 0
bal_acc = (spec + rec) / 2

print(f"\nAccuracy: {test_acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall: {rec:.4f}")
print(f"Specificity: {spec:.4f}")
print(f"F1-score: {f1:.4f}")
print(f"Balanced accuracy: {bal_acc:.4f}")
print(f"Total correct predictions: {np.sum(y_true_test == y_pred_test)}")
print(f"Total incorrect predictions: {np.sum(y_true_test != y_pred_test)}")

# 6. IMAGE VISUALIZATION (Awake vs Sleepy)
plt.figure(figsize=(12, 6))
for i in range(5):
    # Awake example
    awake_idx = np.where(y_true_test == 0)[0][i]
    plt.subplot(2, 5, i+1)
    plt.imshow(test_imgs[awake_idx].astype('uint8'))
    plt.title(f"Class: {classes[0]}")
    plt.axis('off')

    # Sleepy example
    sleepy_idx = np.where(y_true_test == 1)[0][i]
    plt.subplot(2, 5, i+6)
    plt.imshow(test_imgs[sleepy_idx].astype('uint8'))
    plt.title(f"Class: {classes[1]}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```
